```python
apple = fruits[0:100].reshape(-1, 100*100)
pineapple = fruits[100:200].reshape(-1, 100*100)
banana = fruits[200:300].reshape(-1, 100*100)
```

```python
apple.mean(axis=1)
```

```python
np.mean(apple, axis=1)
```

```python
plt.hist(apple.mean(axis=1), alpha=0.8)
plt.hist(pineapple.mean(axis=1), alpha=0.8)
plt.hist(banana.mean(axis=1), alpha=0.8)
plt.legend(['apple', 'pineapple', 'banana'])
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(30, 10))
axes[0].bar(range(10000), apple.mean(axis=0))
axes[1].bar(range(10000), pineapple.mean(axis=0))
axes[2].bar(range(10000), banana.mean(axis=0))
plt.show()
```

```python
fig, axes = plt.subplots(1, 3)
axes[0].imshow(apple.mean(axis=0).reshape(100, 100), cmap='gray_r')
axes[0].axis('off')
axes[1].imshow(pineapple.mean(axis=0).reshape(100, 100), cmap='gray_r')
axes[1].axis('off')
axes[2].imshow(banana.mean(axis=0).reshape(100, 100), cmap='gray_r')
axes[2].axis('off')
plt.show()
```

```python
list_ = [apple, pineapple, banana]
fig, axes = plt.subplots(1, 3)
for i, fruit in enumerate(list_):
  axes[i].imshow(fruit.mean(axis=0).reshape(100, 100), cmap='gray_r')
  axes[i].axis('off')
plt.show()
```

```python
apple_mean = apple.mean(axis=0).reshape(100, 100)
apple_mean.shape
```

```python
pineapple_mean = pineapple.mean(axis=0).reshape(100, 100)
pineapple_mean.shape
```

```python
banana_mean = banana.mean(axis=0).reshape(100, 100)
banana_mean.shape
```

```python
fruits.shape
```

```python
apple_abs_diff = np.abs(fruits - apple_mean)
apple_abs_diff.shape
```

```python
pineapple_abs_diff = np.abs(fruits - pineapple_mean)
pineapple_abs_diff.shape
```

```python
banana_abs_diff = np.abs(fruits - banana_mean)
banana_abs_diff.shape
```

```python
apple_abs_mean = apple_abs_diff.mean(axis=(1, 2))
apple_abs_mean.shape
```

```python
pineapple_abs_mean = pineapple_abs_diff.mean(axis=(1, 2))
pineapple_abs_mean.shape
```

```python
banana_abs_mean = banana_abs_diff.mean(axis=(1, 2))
banana_abs_mean.shape
```

```python
apple_index = np.argsort(apple_abs_mean)[:100]
fig, axes = plt.subplots(10, 10)
for i in range(10):
  for j in range(10):
    axes[i, j].imshow(fruits[apple_index[i*10 + j]], cmap='gray_r')
    axes[i, j].axis('off')
plt.show()
```

```python
pineapple_index = np.argsort(pineapple_abs_mean)[:100]
fig, axes = plt.subplots(10, 10)
for i in range(10):
  for j in range(10):
    axes[i, j].imshow(fruits[pineapple_index[i*10 + j]], cmap='gray_r')
    axes[i, j].axis('off')
plt.show()
```

```python
banana_index = np.argsort(banana_abs_mean)[:100]
fig, axes = plt.subplots(10, 10)
for i in range(10):
  for j in range(10):
    axes[i, j].imshow(fruits[banana_index[i*10 + j]], cmap='gray_r')
    axes[i, j].axis('off')
plt.show()
```
